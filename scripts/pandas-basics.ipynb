```python
from pathlib import Path

# Current notebook location
current = Path.cwd()

# Find the project root (contains pyproject.toml)
root = next(p for p in [current, *current.parents]
            if (p / ".git").exists())


# Import pandas
import pandas as pd

# Load the data from the csv file
csv_path = root / "data" / "orders.csv"
df = pd.read_csv(csv_path)

# Explore the data
print(df)
```

```python
# Print the first 5 rows
print(df.head())

# Print the last 5 rows
print(df.tail())

# Print structural summary: row/column counts, non-null counts per column, dtypes, memory usage — not just dtypes
print(df.info())
```

```python
# Print summary statistics
print(df.describe())

# Print column labels — an Index object native to pandas, not a plain Python list
print(df.columns)

# Print the index metadata
print(df.index)
```
